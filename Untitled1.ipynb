```python
import requests
import time
import pandas as pd
import re
from tqdm.notebook import tqdm
```

```python
def get_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()  # Если ответ в формате JSON
        return data
    print(f"Ошибка: {response.status_code}")
    
def is_valid_string(s):
    pattern = r'^[a-zA-Z!@#$%^&*(),.?":{}|<>]*$'
    return bool(re.match(pattern, s))


vacancies = []
url = 'https://api.hh.ru/vacancies?text="Data science"&per_page=100'
data = get_data(url)
pages = data['pages']
for i in tqdm(range(pages),desc="Обработка страниц, получения URL вакансий"):
    page_url = url+'&page='+str(i)
    page_data = get_data(page_url)
    for j in page_data['items']:
        vacancies.append(j['url'])
print('Найдено',len(vacancies),'вакансий')

words = []
for i in tqdm(vacancies,desc="Обработка вакансий"):
    try:
        response = requests.get(i)
    except ConnectionError:
        print('Соединение сброшено')
    if response.status_code == 200:
        data = response.json()
        words += data['description'].split()
        for i in data["key_skills"]:
            words.append(i['name'])
s = []            
for i in words:
    if is_valid_string(i) and i not in ['and','the','to','of','for','it','a','in','with','or','data','scince'] and len(i)>=2:
        s.append(i.lower().replace(',','').replace('(','').replace(')','').replace('<p>',''))
with pd.option_context("display.max_rows", None, "display.max_columns", None):
    display(pd.DataFrame(s).value_counts())
```
